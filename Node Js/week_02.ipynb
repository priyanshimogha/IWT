Matplotlib and Seaborn Tasks:

1. Plot y=sin⁡ (x) and y=cos⁡ (x) on the same graph for x in [0,2π].

```python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y_sin, label="sin (x)", color="blue")
plt.plot(x, y_cos, label="cos (x)", color="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" sin(x) and cos(x) ")
plt.legend()
plt.grid(True)
plt.show()
```

2. Create a heatmap to show the correlation matrix of numerical columns in the CarPrice_project.csv dataset.

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

plt.figure(figsize=(14, 6))
sb.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".3f")
plt.title("Correlation Matrix - Heatmap")
plt.show()
```

3. Create a combination plot:
● Use a bar plot to show the average price for each carbody.
● Overlay a line plot to show the count of cars for each carbody

```python
avg_price_by_carbody = df.groupby('carbody')['price'].mean()
count_by_carbody = df['carbody'].value_counts()

fig, ax1 = plt.subplots(figsize=(8, 5))
ax1.bar(avg_price_by_carbody.index, avg_price_by_carbody, color='green', label='Average Price')
ax1.set_ylabel('Average Price')
ax1.set_title('Average Price and Count of Cars for each Carbody')

ax2 = ax1.twinx()
ax2.plot(count_by_carbody.index, count_by_carbody, color='red', marker='o', label='Car Count')
ax2.set_ylabel('Count')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```

4. Visualize the relationship between horsepower, enginesize, and price
